# Working with OpenCV & Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('images/bul.jpeg')
rgb_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
print(img.shape)
```

```python
plt.imshow(rgb_image)
plt.axis("off")
plt.show('rgb')
```

```python
crop_image=rgb_image[24:90,50:120,:]
plt.imshow(crop_image)
plt.axis("off")
plt.show('rgb')
```

```python
crop_image.shape
```

```python
total_pixels=crop_image.shape[0]*crop_image.shape[1]
```

```python
print(total_pixels)
```

```python
# Resize the Image
```

```python
crop_image=cv2.resize(crop_image,(20,100))
```

```python
plt.imshow(crop_image)
plt.axis()
plt.title("small image"+str(crop_image.shape))
plt.show('rgb')
```
